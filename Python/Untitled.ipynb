```python
import pandas as pd
from bokeh.io import output_file, show
from bokeh.plotting import figure

df = pd.read_csv("bachelors.csv")

x=df["Year"]
y=df["Engineering"]

output_file("data_bachelors.html")
f=figure()
f.line(x,y)
show(f)
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
import pandas

data = pandas.read_excel("verlegenhuken.xlsx")
temp=data["Temperature"]/10
pres=data["Pressure"]/10

output_file("data_tempres.html")
f=figure()
f.circle(temp, pres, size=0.5)
show(f)

```

```python
from bokeh.plotting import figure, output_file, show
import pandas

data = pandas.read_csv("adbe.csv", parse_dates=["Date"])
f=figure(width=500, height=500, x_axis_type="datetime")
f.line(data["Date"], data["Close"], alpha=0.5)
output_file("Timeseries.html")
show(f)
```
